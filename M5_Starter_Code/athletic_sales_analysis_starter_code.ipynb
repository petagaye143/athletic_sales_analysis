```python
# Import Libraries and Dependencies
import pandas as pd
file_path_2020 = "Resources/athletic_sales_2020.csv"
file_path_2021 = "Resources/athletic_sales_2021.csv"
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020_df = pd.read_csv(file_path_2020)
athletic_sales_2021_df = pd.read_csv(file_path_2021)
```

```python
# Display the 2020 sales Data
athletic_sales_2020_df.head()
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021_df.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_sales_2020_df.info()
```

```python
# Check the 2021 sales data types.
athletic_sales_2021_df.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
merged_inner = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], axis=0, join='inner', ignore_index = True)
merged_inner.reset_index(drop=True, inplace=True)
merged_inner.head(40)
```

```python
# Check if any values are null.
null_count = merged_inner.isnull().sum()
print(null_count)
```

```python
# Check the data type of each column
merged_inner.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
merged_inner['invoice_date'] = pd.to_datetime(merged_inner['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
merged_inner.info()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales
# Show the top 5 results.

result = merged_inner.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index(name= 'Total Sales').sort_values(by='Total Sales', ascending=False)
result.head()
```

```python
# Show the total sales for the products sold for each region, state, and city.
# Optional: Rename the "total_sales" column to "Total Sales"
# Show the top 5 results.
pivot_result = pd.pivot_table(merged_inner, values = 'total_sales', index = ['region', 'state', 'city'], aggfunc='sum').reset_index()
pivot_result = pivot_result.rename(columns={'total_sales':'Total Sales'})
pivot_result = pivot_result.sort_values(by='Total Sales', ascending=False)
pivot_result.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
result =  merged_inner[merged_inner["product"].str.contains("Women's Athletic Footwear")]
result.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
# Show the top 5 results.
womens_footwear_count = result.groupby(['retailer','region', 'state', 'city'])['units_sold'].sum().reset_index(name= 'Womens_Footwear_Units_Sold').sort_values(by='Womens_Footwear_Units_Sold', ascending=False)
womens_footwear_count.head()
```

Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
# Show the top 5 results.
pivot_result = pd.pivot_table(result, values = 'units_sold', index = ['retailer', 'region', 'state', 'city'], aggfunc='sum').reset_index()
pivot_result = pivot_result.rename(columns={'units_sold':'Womens_Footwear_Units_Sold'})
pivot_result = pivot_result.sort_values(by='Womens_Footwear_Units_Sold', ascending=False)
pivot_result.head()
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
# Optional: Rename the "total_sales" column to "Total Sales"
# Show the table.
pivot_result = pd.pivot_table(result, values = 'total_sales', index ='invoice_date', aggfunc='sum')
pivot_result = pivot_result.rename(columns={'total_sales':'Total Sales'})
pivot_result.head()
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
# Sort the resampled pivot table in ascending order on "Total Sales".
pivot_result.index = pd.to_datetime(pivot_result.index)
resample_day = pivot_result.resample('D').sum()
resample_day = resample_day.sort_values(by='Total Sales', ascending=False)
resample_day.head()
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
# Sort the resampled pivot table in ascending order on "Total Sales".
resample_day = pivot_result.resample('W').sum()
resample_day = resample_day.sort_values(by='Total Sales', ascending=False)
resample_day.head()
```
